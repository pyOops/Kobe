# 1. Digital Signals

## 1.1. Definitions

### Analogue signal 模拟信号


**Definitions**: Analogue signal refers to the information expressed by continuously changing physical quantities, such as the image taken by the camera, the pressure, flow rate, speed, humidity, temperature, current, voltage recorded in the workshop control room, also known as **continuous signal, which can have infinite number of different values within a certain time range.**

For nearly a hundred years, whether it is a wired telephone, or wireless transmission of radio and television, use analogue signals to transmit signals. It seems that the analogue signal and the original signal are almost "exactly the same" in the waveform, and it should achieve good propagation effect. However, the fact is quite the opposite. 

For radio stations, there is always a greater lack of effect. There are also glimmerings of snowflakes on TV images. This is because the signal in the transmission process to go through a lot of processing and forwarding, these devices will inevitably produce some interference.

In addition, if it is wired transmission, electrical equipment near the line also generates electromagnetic interference(电磁干扰）.If it is wireless transmission, it is more "open", and all kinds of interference in the air can not be resisted.

These interferences can easily cause signal distortion and also bring some noise. These distortions and additional noise will also accumulate as the transmission distance increases, seriously affecting the communication quality.

### Digital  signal数字信号

**Definitions**: Digital signal refers to the signal that is discrete and discontinuous in value. Digital signals are formed by sampling, quantization and coding on the basis of analogue signals. Specifically,

**Sampling**: is to get the sample value of each time moment by the appropriate time interval of the input analogue signal. 


**Quantization**: is to use the binary code system to express the value of each moment measured by sampling in order to be acceptable to the computer. 


**Coding**: is to arrange the binary numbers generated by quantization to form a sequential pulse sequence.


###  Why transfer analogue signal to digital  signal

**Analogue signal**: because of its variability, so it is very easy to suffer from the interference of the external environment, there are countless types of waveforms, and a waveform as long as a slight change, it will become another completely different waveform.

**Digital signal**: limited types of waveforms ensure that it has strong anti-interference ability, and as long as the disturbed waveform does not exceed a certain limit, it can always be recovered by some shaping circuits（整形电路）, which ensures the accuracy and credibility of the signal.


###   Conversion process

The conversion process between analogue signal and digital signal is mainly carried out in the sensor, in the whole conversion process, the sensor will do filtering (滤波） and other operations, after that, the analog signal will be converted into digital signal. Then transmitted out through wired or wireless means. After the electrical signal is received, it is restored into an information signal through the receiving device.

## 1.2. Sampling and Quantiztion

In this chapter you will look at time based signals. In order for an analogue signal（模拟信号） to be used in the computer, the signals are sampled from a continuous analogue signal. That is, convert an analogue signal into a digital one.This involved two processes:


Continuous function：$$ f (x)=1+2x$$  
Discrete series：$$f (n)=1+2n$$

1) **Sampling**: An A to D converter reads an analog signal at regular time intervals, measuring the value of the signal at that point. Each reading is called a **sample（样本）**.

*A-D converter（模数转换器）: an electronic component that converts an analog signal into a digital one.*

2) **Quantization**:The measured value is assigned to the nearest digital value.

The sampling signal turns the digital signal by integer quantization, that is, after the process of rounding（整数化处理）. 将每个采样点得到的幅值以数字存储。

*The amplitude of the sampled signal is infinite. The continuous sampled signal with infinite amplitude are passed through a quantizer and rounded to a finite number of **M amplitudes (M is an integer)**. This is quantization.  Since the amplitude is integer in the quantization process, there is a difference between the quantized signal and the sampled signal, **called quantization error**, so that there is a certain error between the signal at the receiver and the original signal. This error is represented as the received noise, which is called **quantization noise**. The finer M, the smaller the error, and the smaller the quantization noise.*

These two processes are shown in the figure below, where a pseudo （/ˈsuːdəʊ/）analogue signal (A digital signal sampled at a high sample rate to represent a continuous analogue signal) is sampled at **regular time steps (sample rate)** and then these values are assigned to the nearest 0.25 value.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
# The following lines change the default setting for the graphs.
plt.rcParams['axes.labelsize'] = 22  # 轴旁Time
plt.rcParams['xtick.labelsize'] = 16  # 轴旁数字
plt.rcParams['ytick.labelsize'] = 16
plt.rcParams['legend.fontsize'] = 16 # 标题大小
plt.rcParams['axes.titlesize']= 22  #title
%matplotlib inline

#假定一个模拟信号（事实上是一个高样本率的数字信号，代表连续的模拟信号）
SR = 100  
time = np.arange(0,5+1/SR,1/SR) #获得0-5的时间线
signal = 1.8*np.sin(np.pi*time)

#取样率
sampleRate =30  #change this 1s内采15个样本 
#量化位数
bits = 8 #change this from 1 to 8 or 16, n is qantitative number  
level = 2/(2**bits)  #  the nth power of 2 (2^n),qantitative level 
##取样信号
sampleTime = np.arange(0,5+(1/sampleRate),1/sampleRate)  
f = interp1d(time,signal,kind='cubic')
sampleSignal = f(sampleTime)  
##量化信号
quantSignal = np.round((sampleSignal/level),0)*level  # The sample signal is magnified by one levelths(1/level) and rounded
#样本信号被放大的了1/level倍，进行四舍五入

plt.figure(figsize=(15,8))
plt.plot(time,signal,'k-', label = 'Analogue Signal')
plt.stem(sampleTime,sampleSignal,label = 'Sampled Signal',linefmt='r--',markerfmt='ro')
plt.plot(sampleTime,quantSignal,'b.-',label='Quantized Signal')
plt.legend(loc ='upper right')
plt.xlabel('Time [s]')
plt.ylabel('Signal Amplitude')
plt.grid()
```

# 2 Aliasing-SampleRate

Aliasing occurs when signals are sampled too slowly. If the sample rate is too slow, the data will not accurately represent the signal content. Worse, it will create a spurious （/ˈspjʊəriəs/ faulse） signal which is not really there, called an alias（混叠/ˈeɪliəs/）.

The GREEN lines show a 100Hz signal sampled at 1000 samples/sec.The RED lines show a 900Hz signal sampled at 1000 samples/sec The dotted line is an alias signal, an apparent signal at 100Hz, which did not exist in the original signal. Once acquired, there is no method of distinguishing the true signal from the alias. 

**The Nyquist Sampling Criterion states that the sample rate must be at least twice （maybe 10-20 times） the highest frequency present in the signal, otherwise aliasing will occur.**

Anti-aliasing filters are used when acquiring digital signals to prevent this problem

Try varying the frequency of **signal2** from 900 to 100 Hz to see what happens when your signal drops below the Nyquist frequency.

```python
# analogue signal
SR = 100000
time = np.arange(0,0.01+1/SR,1/SR)
signal1 = np.sin(2*np.pi*100*time)
signal2 = np.sin(2*np.pi*900*time) #Alter this

# digital signal
sampleRate = 10000  #  1s 中取1000个，0.01s取10个。if this number is twice of the frequency, that has two sample in one circle
sampleTime = np.arange(0,0.01+(1/sampleRate),1/sampleRate)  # the highest frequency is 100, 
f1 = interp1d(time,signal1,kind='cubic')
f2 = interp1d(time,signal2,kind='cubic')
sampleSignal1 = f1(sampleTime)
sampleSignal2 = f2(sampleTime)

plt.figure(figsize=(15,6))
ax1 = plt.subplot(1,2,1)
plt.plot(time,signal1,'k-', label = 'Analogue Signal')  # 模拟信号
plt.stem(sampleTime,sampleSignal1,markerfmt='go',linefmt='g-')
plt.plot(sampleTime,sampleSignal1,'r--')
plt.xlim(0,0.01)  #plt.ylim() 显示的是y轴的作图范围
plt.xlabel('Time [s]')
plt.ylabel('Signal Amplitude')
plt.grid()
plt.legend()

ax1 = plt.subplot(1,2,2)
plt.plot(time,signal2,'k-', label = 'Analogue Signal')
plt.xlabel('Time [s]')
plt.ylabel('Signal Amplitude')
plt.grid()
plt.xlim(0,0.01)
plt.stem(sampleTime,sampleSignal2,markerfmt='ro',linefmt='r-')
plt.plot(sampleTime,sampleSignal2,'r--')
plt.legend()
```

## 3.2 Artifacts（震荡）

Experimental data is usually not perfect and has a number of artifacts （震荡）which can detract from the original data. **These could be potentially from sources of electrical noise, faults in the saving of the data, drift in the sensor etc （电子噪声、数据保存中的故障、传感器漂移）**.  


Sensor drift refers to the phenomenon that the sensor output changes with time when the input quantity is unchanged. There are two reasons for the drift: 

1. Structural parameters of the sensor itself; 
2. Surrounding environment (such as temperature, humidity, etc.). 

The most common drift is temperature drift, that is, the change of output caused by the change of ambient temperature. The sensor detects some external parameters, and then returns some analogue and digital signals, due to some interference from the external environment, etc., resulting in the signal returned by the sensor is mixed with some interference signals, because the sensor needs to send the signal back, the signal needs to be amplified (because the unprocessed signal is small), but the correct signal amplification also amplifies the interference signal.

Perhaps at the beginning of the interference signal is relatively large, after amplification, interference signal has occupied the main signal. The original correct signal is completely changed into a signal dominated by the interference signal. At this time, the signal transmitted by the sensor has changed, that is, drift.

In order to simulate the typical signal corruption in a controlled manner we now corrupt the signal with a number of sources of noise. The example below takes the original signal and adds the following sources of noise:
1. Random noise
2. Linear drift
3. A signal offset
4. Noise at 250 Hz
5. Noise at 15 Hz

```python
random_noise = Series(0.1*np.random.randn(t.size), index=t)  # random noise 随机噪音
# np.random.randn() return of array for np.size()-Standard normal distribution-0 was taken as the mean and 1 as the standard deviation,N（0，1）
# np.random.rand() return of array for np.size()--uniform distribution-random sample value from 0 to 1 for np.size() ,[0,1)
drift = Series(1*t, index=t)                                 # Linear trend  线性趋势
offset = Series(0.5, index=t)                                # Offset the mean 信号位相差 信号偏移
noise250 = Series(0.1*np.sin(2*np.pi*250*t), index=t)        # Noise at 250Hz  噪音
noise15 = Series(0.2*np.sin(2*np.pi*15*t), index=t)          # Noise at 15Hz
zc = z + random_noise + drift + offset + noise250 + noise15

plt.figure(figsize=[20,10])
plt.plot(z,'b')
plt.plot(zc,'r')
plt.xlabel('time [s]')
plt.ylabel('Signal z(t)')
plt.legend(['Original z','Corrupted zc'])
plt.grid()
```

Visually, you should be able to see that the original (blue) signal and the corrupted (red) signal diverge over time. You can see from the mean and maximum values of the original and corrupted signals below that they are quite different. Beyond that it is difficult to see the difference between the two signals.

```python
print('Mean of original signal = ',round(z.mean(),3),'\n')
print('Mean of corrupted signal = ',round(zc.mean(),3),'\n')
print('Max of original signal =',round(z.max(),3),'\n')
print('Max of Corrupted signal =', round(zc.max(),3),'\n')
print('Original Signal Properties =\n',z.describe(),'\n')
print('corrupt Signal Properties =\n',zc.describe(),'\n')
```

## 4.1 Analyse the frequency content of the signal

In digital communication, analogue signals should be sampled and transmitted. In this way, in the time domain, no matter how intensive the instant sampling period is, the original signal cannot be uniquely restored. That is, there is a loss of information in the signal in the time domain. However, in frequency domain, the sampled signal is the shift and accumulation of the original signal in the frequency domain. If the sampling method satisfies the sampling theorem, that is, the sampling frequency is twice the highest frequency of the original signal (oversampling), the baseband frequency domain response of the original signal can be obtained through the low-pass filter. The original signal is then obtained using the inverse Fourier transform. It can be seen that a signal with information loss after sampling can be completely recovered by Fourier transform, low-pass filter shaping, and inverse Fourier transform. This is where the time-frequency variation comes in.

It is difficult to identify the frequency components by looking at the original signal in the time domain, it needs to be converted in to the frequency domain. The discrete Fourier transform of the noisy signal Y is found by taking the fast Fourier transform (FFT) of the time signal. 

You should have covered Fourier series in your maths modules, but briefly a Fourier series recreates a signal over a fixed period by the summation of a finite number of sine and cosine waves. 
$$ f(t) = \frac{1}{2} a_{0} + \sum_{n=1}^{\infty} a_n cos(n,t) + \sum_{n=1}^{\infty} b_n sin(n,t)$$

Fourier analysis determines the amplitude of these component sine and cosine waves. The Fast Fourier Transform (FFT) is a special case of fourier analysis which uses efficient routines provided the length of the signal is $2^n$.

First look at the help on the fft function.

The Fourier PRINCIPLE states that any time sequence signal measured continuously can be expressed as an infinite superposition of cosine (or sine) wave signals of different frequencies.  FFT is a fast algorithm for discrete Fourier transform, which can transform a signal into the frequency domain.  So what are its uses in practice?  

1) Some signals in the time domain is difficult to see the characteristics, but if the transformation to the frequency domain, it is easy to see the characteristics (frequency, amplitude, initial phase);  

2) FFT can extract the spectrum of a signal for spectrum analysis and preparation for subsequent filtering;  

3) A preliminary understanding of the system can be obtained by comparing the input signal and output signal of a system with the Fast Fourier Transform.  

Suppose the sampling frequency Fs, signal frequency F, signal length L, and sampling number N.  So the result after FFT is a complex number (复数）of N points.  Each of these points corresponds to a frequency point.  The magnitude of this point is the amplitude characteristic at that frequency.

The code below, calculates the FFT of the timeseries **zc** and plots the double-sided amplitude spectrum $|Y|$ against the frequency in Hz. 

If you plot the complete FFT you will see that it is symmetrical about the Nyquist frequency of 500Hz. You will also see spikes（峰值） in the curve corresponding to the main frequency components of 120 Hz and 50 Hz. There is also a large response at 0 Hz.

```python
from pandas import DataFrame, Series
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

#help(pd.Series)

Fs = 1000 # Sample Frequency in Hz
T = 1/Fs # Sample Time step
E = 10 # end time
t = np.arange(0,E,1/Fs) # Time Vector
L = np.size(t)


z1 = Series(0.7 * np.sin (2*np.pi*120*t), index=t) #The Series function creates a pandas series from an array like object
z2 = Series(1 * np.sin (2*np.pi*50*t), index=t) #creates onther series
z = z1 + z2 # Adds two series together

plt.figure(figsize=[20,10])
plt.plot(z,'b')
plt.plot(zc,'r')
plt.xlabel('time [s]') 
plt.ylabel('Signal z(t)')
plt.legend(['Original z','Corrupted zc'])
plt.grid()

random_noise = Series(0.1*np.random.randn(t.size), index=t)  # random noise 随机噪音
drift = Series(1*t, index=t)                                 # Linear trend  线性趋势
offset = Series(0.5, index=t)                                # Offset the mean 抵消平均值
noise250 = Series(0.1*np.sin(2*np.pi*250*t), index=t)        # Noise at 250Hz  噪音
noise15 = Series(0.2*np.sin(2*np.pi*15*t), index=t)          # Noise at 15Hz

zc = z + random_noise + drift + offset + noise250 + noise15

# 之前的代码
nfft = 2**(int(np.ceil(np.log2(L))))  # find next power of 2 longer than the length of the signal 
# int(np.ceil) 对元素向上取整  # np.log2(L) 以2为底的对数
f = Fs*np.linspace(0,1,nfft) # calculate frequency
Y = Series(np.fft.fft(zc,n=nfft)/L, index=f)     # fft 得到一组复数数组

print('L = ',L,'\n')
print('nfft = ',nfft,'\n')
print('f = ',f,'\n')
print('Y = ',Y,'\n')

plt.figure(figsize=[20,10])
plt.plot(Y[0:Fs].abs()) # plot tha absolute value of the amplitude 从0-Fs
plt.xlabel('Frequency [Hz]')
plt.ylabel('|Y|')
plt.title('Double Sided Amplitude Spectrum')
plt.grid()
```

It is better to plot the single sided amplitude spectrum because the double sided spectrum is symmetrical about the Nyquist frequency. 

The Nyquist theorem states that the sampling frequency must be at least twice the maximum frequency of the signal we are measuring to avoid aliasing. For example if we have a signal which has a maximum frequency component of 100 Hz, we must sample it at a frequency of at least 200 Hz. In the case below we are sampling at 1000 Hz, so the maximum frequency component of the signal can be 500Hz to avoid aliasing.


```python
plt.figure(figsize=[10,5])
plt.plot(2*(Y[0:Fs/2].abs())) #plot the amplitude spectrum 2*abs(Y) against frequency from 0 to Fs/2 (Nyquist Frequency)
plt.xlabel('Frequency [Hz]')
plt.ylabel('2*|Y|')
plt.title('Single Sided Amplitude Spectrum')
plt.grid()
```

Because the x-axis is quite spread out, it is quite common to use a log axis.
help(plt.semilogx)

```python
plt.figure(figsize=[10,5])
plt.semilogx(2*(Y[0:Fs/2].abs())) # plot the amplitude spectrum 2*abs(Y) against frequency from 0 to Fs/2 (Nyquist Frequency)
plt.xlabel('log(Frequency) [Hz]')
plt.ylabel('2*|Y|')
plt.title('Single Sided Amplitude Spectrum on Log-x axis')
plt.grid(which='both')
```

## 4.2 Detrend the signal
You will see from the figure above that there is a large low frequency component which is dominating the signal. In this case it is caused by the linear drift and offset we introduced to the signal. So we need to remove this using the scipy.signal **detrend** command. Using the linear type, the detrend command will remove the linear increase with time.


```python
from scipy import signal
zcd = Series(signal.detrend(zc, type=='linear'), index=t)
plt.figure(figsize=[20,10])
plt.plot(zc[0:10])
plt.plot(zcd[0:10])
#plt.plot(zcd[0:5])
plt.title('Detrended Corrupted Signal')
plt.xlabel('time')
plt.legend(['Corrupt','Detrend'])
plt.grid()
```

Now re-calculate the FFT of the detrended signal to show the frequency components.

```python
nfft = 2**(int(np.ceil(np.log2(L))))   # find next power of 2
f = Fs*np.linspace(0,1,nfft) # calculate frequency
Y = Series(np.fft.fft(zcd,n=nfft)/L, index=f)                    # fft
              
plt.figure(figsize=[20,10])
plt.semilogx(2*(Y[0:Fs/2].abs())) # 2*abs(Y) because it is the single-sided amplitude
plt.xlabel('log(Frequency) [Hz]')
plt.ylabel('2*|Y|')
plt.grid(which='both')
```

You should now be able to clearly see four distinct peaks in the amplitude spectrum at about 15Hz, 50Hz, 120Hz and 250Hz. The main reason the amplitudes do not exactly match the original amplitudes of 0.7 and 1 for the two main frequency components is because of the noise. Several executions of this code (including recomputation of Y) will produce different approximations to 0.7 and 1. The other reason is that you have a finite length signal. Increasing L from 1000 to 10000 in the example above will produce much better approximations on average.

## 4.3 Filters


Filters are used to remove unwanted effects from the signal.

- High-pass filters remove frequencies below the cut-off frequency $f_c$ and allow higher frequencies to pass.

- Low-pass filters remove frequencies above the cut-off frequency $f_c$ and allow lower frequencies to pass and are used to reduce noise, which tends to be predominantly at high frequencies.

- Band-pass filters preferentially pass frequencies of interest.

- Band-stop filters reduce the effect of particular frequencies such as mains hum.

All filters modify the signal. Ensure that the modification is acceptable.

There is still a peak at 15Hz and 250Hz which is not related to the original signal. This is due to the noise we added to the original signal at these two specific frequencies. In order to remove this noise we will need to use filters. We can use a highpass filter to remove the noise at low frequencies, below 20Hz and a lowpass filter to remove the noise above 150Hz. 

The low pass filter will allow frequency components below the cut off frequency to pass and attenuate the frequencies above the cut off frequency. Whereas the highpass filter will allow frequencies above the cut off frequency to pass and attenuate the frequencies below the cut off. The design of filters is a complicated area which we won't go into in detail on this course. But you can see from the two plots below how the filter response fits to the amplitude spectrum.

The first subplot show both a lowpass filter (green dotted) and a highpass filter (red dash) and the amplitude spectrum (blue)

The second subplot shows a bandpass filter (red) and the amplitude spectrum (blue).

What happens to the filter response as you change the filter frequencies and the number of poles, npoles between 2 and 10.

```python
SR = Fs=1000 # Sample Frequency in Hz

npoles = 10
f_low = 140
f_high = 35
# Design Butterworth filter
b_low, a_low = signal.butter(npoles, f_low/(SR/2), 'low', analog=False, output='ba')
b_high, a_high = signal.butter(npoles, f_high/(SR/2), 'high', analog=False, output='ba')
b_band, a_band = signal.butter(npoles, (f_high/(SR/2), f_low/(SR/2)), 'bandpass', analog=False, output='ba')

# Determine frequency response of filter
w_low, h_low = signal.freqz(b_low, a_low)
w_high, h_high = signal.freqz(b_high, a_high)
w_band, h_band = signal.freqz(b_band, a_band)

# Dimensionalise the frequencies
w_low = w_low/(2*np.pi)*Fs
w_high = w_high/(2*np.pi)*Fs
w_band = w_band/(2*np.pi)*Fs

plt.figure(figsize=(16,7))
ax1 = plt.subplot(1,2,1)
plt.plot(w_low, abs(np.array(h_low)),'g:', label = 'lowpass')
plt.plot(w_high, abs(np.array(h_high)),'r--', label = 'highpass')
plt.title('Low-Pass Frequency Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.grid(which='both', axis='both')
plt.plot(2*(Y[0:Fs/2].abs()), label = 'Signal')
plt.legend()

ax2 = plt.subplot(1,2,2)
plt.plot(w_band, abs(np.array(h_band)),'r-', label = 'band-pass')
#plt.xscale('log')
plt.title('Band-Pass Frequency Response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
#plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.plot(2*(Y[0:Fs/2].abs()), label = 'Signal')
plt.legend()
```

Having designed a suitable filter, you now need to apply the filter to the signal. This is done in the time domain. You can use either `lfilter` which applies the butterworth filter（巴特沃斯滤波器） designed by the `butter` command to a signal **in one direction**. But this can introduce a phase lag（相位滞后）, see the first figure below. 

Or use the `filtfilt` command which runs the signal through the filter in both direction which removed any phase lag, see second figure below.

```python
# Filter the data using lfilter
zcdf = Series(signal.lfilter(b_band,a_band,zcd), index=t) # produces a phase lag

plt.figure(figsize=[20,10])
plt.plot(z[0:0.1])
plt.plot(zcdf[0:0.1])
plt.legend(['Original','filtered'])
plt.title('Comparison of original and low pass filtered signal using lfilter')
plt.grid()
```

```python
# Filter the data using filtfilt
zcdf = Series(signal.filtfilt(b_band,a_band,zcd), index=t) # filtfilt runs the filter in both directions to remove phase changes. 

plt.figure(figsize=[20,10])
plt.plot(z[0:0.1])
plt.plot(zcdf[0:0.1])
plt.legend(['Original','filtered'])
plt.title('Comparison of original and low pass filtered signal using filtfilt')
plt.grid()
```

You should be able to see from above, the band pass filtered signal nearly matches the original signal.
