## 1 Fourier Transform

### 1.1 Definition

The Fourier transform represents the ability to express a function satisfying certain conditions as a trigonometric function (sine and/or cosine) or a linear combination of their integrals.
In different fields of study, the Fourier transform has many different variants, such as the continuous Fourier transform and the discrete Fourier transform. Fourier's analysis was originally proposed as a tool for analytical analysis of thermal processes.

### 1.2 Orthogonality of Function

#### Family of trigonometric functions

The trigonometric function family consists mainly of the following family of functions, all of which are sine or cosine functions:
$$
\{
1, \sin x, \cos x, \sin 2x, \cos 2x, ..., \sin nx, \cos nx
\}
$$
#### Orthogonality

Orthogonality is the concept of linear algebra, a generalization of the intuitive concept of vertical. As an adjective, it has meaning only in a definite inner product space. Two vectors in the inner product space are said to be orthogonal if their inner product is $0$. Orthogonality can be intuitively understood as vertical if the Angle between vectors can be defined. In physics: the independence of motion can also be explained in terms of orthogonality. In addition, the definition of orthogonal function system is supplemented here: in the trigonometric function system, the integral of any product of two different functions on the interval $[-\pi,\pi]$ is equal to $0$, then the system of such trigonometric functions is called the orthogonal function system.

For example, the following $2$ functions can be said to be orthogonal:
$$
f(x)= \sin x \quad \text{and} \quad g(x)=\cos 2x
$$
because:
$$
\int_{-\pi}^{\pi} f(x)·g(x)\text{d}x =0
$$

As a theorem, the following four equations will hold, which will prove the orthogonality between them, where $n$ and $m$ are both intergers:
$$
\int_{-\pi}^{\pi}\sin nx \cos mx \text{d}x=0 \quad (n\ne m)
$$

$$
\int_{-\pi}^{\pi}\cos nx \cos mx \text{d}x=0 \quad (n\ne m)
$$

$$
\int_{-\pi}^{\pi}\cos nx \sin mx \text{d}x=0 \quad (n\ne m)
$$

$$
\int_{-\pi}^{\pi}\sin nx \sin mx \text{d}x=0 \quad (n\ne m)
$$

### 1.3 Fourier expansion of a function with period of $2\pi$

For any periodic function, we can expand it as follows:
$$
f(x)=\frac{a_0}{2}+\sum_{n=1}^{\infty}(a_n\cos nx+b_n\sin nx) \tag{1}
$$
where:
$$
a_0=\frac{1}{\pi}\int_{-\pi}^\pi f(x) \ \text{d}x
$$

$$
a_n=\frac{1}{\pi}\int_{-\pi}^\pi f(x)\cos nx \ \text{d}x
$$

$$
b_n=\frac{1}{\pi}\int_{-\pi}^\pi f(x)\sin nx \ \text{d}x
$$

### 1.4 Fourier expansion of a function with period of $2L$

If we extend the above case of fixed period $2\pi$ to the more general case, that is, period $2L$, and this $L$ can be any real number, what should we do?

In fact, the results are very similar, and we can still convert these periodic functions to a fixed-style Fourier series:
$$
f(t)=\frac{a_0}{2}+\sum_{n=1}^\infty\bigg[a_n\cos n\omega t+b_n\sin n\omega t\bigg] \tag{2}
$$
where:
$$
a_0=\frac{2}{T}\int_{0}^T f(t) \ \text{d}t
$$

$$
a_n=\frac{2}{T}\int_0^T f(t)\cos n\omega t \ \text{d}t
$$

$$
b_n=\frac{2}{T}\int_0^T f(t)\sin n\omega t \ \text{d}t
$$

$$
T=2L
$$

### 1.5 The complex form of Fourier series

It is well known that both sine and cosine functions can be expressed in complex numbers:
$$
\cos\theta=\frac{e^{i\theta}+e^{-i\theta}}{2}
$$

$$
\sin\theta=\frac{e^{i\theta}-e^{-i\theta}}{2i}
$$

Then, plug these contents into formula $(2)$, and after a simplification, we get:

$$
f(x)=\sum_{n=-\infty}^{+\infty}c_n e^{in\omega t} \tag{3}
$$

Formula $(3)$ is called the complex form of the Fourier series, and in fact, in the Fourier transform we will discuss later, the complex form of the Fourier series expression will be more common.

### 1.6 Fourier Transform
The Fourier transform is a generalization of the Fourier series that decomposes aperiodic functions into the sum of a series of sine and cosine functions. The formula for the Fourier transform is as follows:

$$
\mathcal{F}(\omega)=\int f(t)·e^{-i\omega t}·\text{d}t \tag{4}
$$

Using Fourier transform, we can easily convert **time domain** signals into **frequency domain** signals. In this way, we can easily see how much of the "contribution" of different frequency subsignals of a total signal is, that is, how large the amplitude is.

> We call the function $\mathcal{F}(\omega)$ is the frequency domain signal function, $f(t)$ is the time domain signal function.

In formula $(4)$, $f(t)$ is an aperiodic function, $\mathcal{F}(\omega)$ is the representation of the function in the frequency domain, $e^{-i\omega t}$ is the complex exponential function, and $\omega$ is the angular frequency.

The physical meaning of the Fourier transform is that <b><u>any aperiodic function can be expressed as the sum of sine and cosine functions of many different frequencies</u></b>. These sine and cosine functions are called the basis frequency, and the frequency of the basis frequency is continuous and can take any real value.

### 1.7 Properties of the Fourier Transform
The symbol conventions in this section are as follows:

- $\mathscr{F}[f(t)]$: Take the Fourier transform of $f(t)$
- $\mathcal{F}(\omega)$: The frequency domain function of $f(t)$
#### 1.7.1 Linear property
Suppose $\mathscr{F}[f(t)]=\mathcal{F}(\omega)$, $\mathscr{F}[g(t)]=\mathcal{G}(\omega)$, $\alpha$ and $\beta$ are constants, then:

$$
\mathscr{F}[\alpha f(t)+\beta g(t)]=\alpha \mathscr{F}(\omega)+\beta\mathcal{G}(\omega)
$$

The essence of Fourier transform is that the original function (time domain) is represented linearly by various periodic functions with different frequencies (frequency domain), which must have linearity. This is consistent with the linearity of the integral.

#### 1.7.2 Displacement Property

Suppose $\mathscr{F}[f(t)]=\mathcal{F}(\omega)$, $t_0$ and $\omega_0$ are constants, then,

$$
\mathscr{F}[f(t-t_0)]=e^{-i\omega t_0}\mathcal{F}(\omega)
$$

The function in the time domain is shifted $t_0$ to the right, which is equivalent to the time starting point is changed to $-t_0$, and the phase in the frequency domain is also returned accordingly. The value of the component $e^{i\omega t}$ at the value $e^{-i\omega t_0}$ at time $t_0$ is the starting point, so $\mathcal{F}(\omega)$ is multiplied by the quantity.

#### 1.7.3 Shrinkage Property

Suppose $\mathscr{F}[f(t)]=\mathcal{F}(\omega)$, $a$ is a non-zero constant, then

$$
\mathscr{F}[f(at)]=\frac{1}{|a|}\mathcal{F}\bigg(\frac{\omega}{a}\bigg)
$$

This operation obviously speeds up the frequency of each component by a factor of $a$. To counteract this change, divide $\omega$ by $a$, but at the same time the coefficients of the components also change proportionally, as shown in the figure below:

![Intuitively understand the deflating properties of the Fourier transform](https://pic3.zhimg.com/80/v2-d813375c4c8e730dd62907dc25e6b2c6_1440w.webp)

#### 1.7.4 Differential Relation

Suppose $\mathscr{F}[f(t)]=\mathcal{F}(\omega)$, as long as the relevant derivative exists, then:

$$
\mathscr{F}\bigg[\frac{\text{d}^n f(t)}{\text{d}t^n}\bigg]=
(i\omega)^n \mathcal{F}(\omega)
$$

If the function $f(t)$ being transformed is a complex-valued function, the meaning of its derivative is complex-valued velocity, and the components of the velocity along the real and imaginary axes represent the real and imaginary velocities, respectively.

![](https://pic2.zhimg.com/80/v2-d82e6785af4ac8d78fbc44199e4a1b61_1440w.webp)

#### 1.7.5 Integral Relation

Suppose $g'(t)=f(t)$, $\mathscr{F}[f(t)]=\mathcal{F}(\omega)$, then

$$
\mathscr{F}[g(t)]=\frac{1}{i\omega}\mathcal{F}(\omega)
$$

This is consistent with the differential relation, by substituting $n$ for $-1$.

## 2 Fast Fourier Transform
### 2.1 What is Fast Fourier Transform (FFT)?

Due to the rapid development of computer technology, in the mid-1970s, some electronic equipment companies in the United States and Japan began to design and produce digital Fourier analyzer, but because of the large amount of computation of discrete Fourier transform, until the fast algorithm of DFT (FFT) was discovered, finite discrete Fourier transform (DFT) was really widely used. FFT is an efficient algorithm of DFT called fast Fourier transform. Fourier transform is one of the most basic methods in time-frequency domain transform analysis. The Discrete Fourier Transform (DFT) used in digital processing is the basis for many digital signal processing methods.

### 2.2 The difference between a fast Fourier transform and an ordinary Fourier transform
The Fast Fourier Transform allows us to compute some things with lower time complexity, and the following is an example to illustrate this conclusion.

For polynomial $f(x)=\sum_{i=0}^n a_ix^i$, $g(x)=\sum_{i=0}^n b_i x^i$, define its product as $fg$ as:

$$
(fg)(x)=\Bigg(\sum_{i=0}^n a_ix^i\Bigg)\Bigg(\sum_{i=0}^n b_i x^i\Bigg)
$$

Based on common sense in computer science, it is obvious that the coefficients of each of the above terms are computed using a two-layer traversal iteration, that is, the time complexity is $\text{O}(n^2)$, but the fast Fourier transform allows us to compute the coefficients in a time complexity of $\text{O}(n\log n)$.

### 2.3 FFT

For polynomial $f(x)=\sum_{i=0}^{n-1} a_ix^i \in X_{n-1}$, without losing generality, let $n=2^s \quad s\in\mathscr{N}$ (since in multiplication of polynomials we can think of a polynomial equivalent to a polynomial whose coefficients of higher degree are all zero, so we can think of $n$ as the first integer power greater than or equal to its $2$), consider the parity of $a_i$subscript to divide the term $f(x)$ into two parts, namely:

$$
f(x)=(a_0+a_2x^2+a_4x^4+\dots+a_{n-2}x^{n-2}) + 
(a_0+a_2x^2+a_4x^4+\dots+a_{n-1}x^{n-1})
$$

Let

$$
f_1(x)=a_0+a_2x^1+a_4x^2+\dots+a_{n-2}x^{\frac{x}{2}-1} \\
f_2(x)=a_1+a_3x^1+a_5x^2+\dots+a_{n-1}x^{\frac{x}{2}-1} 
$$

Then

$$
f(x)=f_x(x^2)+xf_2(x^2)
$$

Plug in $x=\omega_n^k \quad (k<\frac{n}{2})$:

$$
\begin{aligned}
f(\omega_n^k)&=f_1(\omega_n^2k)+\omega_n^k f_2(\omega_n^2k) \\ 
&=f_1\big(\omega_{\frac{n}{2}}^k\big)+\omega_n^k f_2\big(\omega_{\frac{n}{2}}^k\big)
\end{aligned}
$$

Plug in $\omega_n^{k+\frac{n}{2}} \quad (k<\frac{n}{2})$:

$$
\begin{aligned}
f(\omega_n^{k+\frac{n}{2}})&=f_1(\omega_n^{2k+n})+\omega_{n}^{k+\frac{n}{2}}f_2(\omega_n^{2k+n}) \\
&=f_1(\omega_n^{2k}·\omega_n^n)-\omega_n^kf_2(\omega_n^{2k}) \\
&=f_1(\omega_n^{2k}-\omega_n^k f_2(\omega_n^{2k})) \\
&=f_1(\omega_{\frac{n}{2}^k})-\omega_n^k f_2(\omega_{\frac{n}{2}}^k)
\end{aligned}
$$

As a result:
$$
f_1(\omega_1^k)=f_2(\omega_1^k)=1
$$

So you can find all the coefficients of f in the time complexity of $\text{O}(n\log n)$
