## 0 What is Ship Maneuverability?
The maneuverability of a ship refers to the ability of the ship to be manipulated and controlled by the pilot during the voyage.

## 1 General equation of plane maneuvering motion

### 1.1 Preknowledge

Before you can understand the content of this article, you need to have some basic mathematical knowledge, such as vector calculus, linear algebra, and fluid dynamics. Below, we'll review each of these.

#### 1.1.1 Vector

Ships travel in water, so we have to understand fluid dynamics. One of the most common things in fluid dynamics is vectors. We call a quantity that has both magnitude and direction a <b style="color:red">VECTOR</b>. For example, force, momentum, etc., are vectors.

#### 1.1.2Basic Operations of Vectors

Like scalars, vectors can be added, subtracted, multiplied, and divided. However, since the vector has a direction, the calculation of addition, subtraction, multiplication and division must be performed not only on the size but also on the direction. In general, in a coordinate system, a vector can be determined by multiple scalars that form a coordinate, and the "direction" formed by the coordinate and the origin is called the direction. Therefore, we can express the vector in coordinate form, and then add, subtract, multiply, and divide four operations for each component (since these coordinates are all scalars), and finally get a new vector.

In the geometric sense, the addition or subtraction of a vector means that the head of one vector (arrow position) is connected to the tail of another vector, and the superimposed vector is equal to the tail of the first vector pointing to the head of the second vector, which is the new vector.

Multiplication of vectors is mainly a cross product and a dot product. This is due to the direction. According to the mathematical definition:

$$
\vec{A}·\vec{B}=|\vec{A}||\vec{B}|\cos\theta
$$

For cross product, then:

$$
|\vec{A}\times\vec{B}|=|\vec{A}||\vec{B}|\sin\theta
$$

$$
\vec{A}\times\vec{B}=
\begin{vmatrix}
\vec{i} & \vec{j} & \vec{k} \\
A_x & A_y & A_z \\
B_x & B_y & B_z
\end{vmatrix}
$$

,where $\theta$ represents the angle between two vectors (generally take the smaller angle), $A_x, A_y$ and $A_z$ represents components of vector $\vec{A}$, and, $B_x, B_y$ and $B_z$ represents components of vector $\vec{B}$.

### 1.2 Introduction

Imagine you're driving a submarine. A submarine is like a car running under water. When you want to change the direction of the submarine, you turn something called a "rudder," just like you turn the steering wheel when you drive a car.

When you turn the rudder, it creates a current behind the submarine that acts like a hand, pushing the submarine in the direction you want it to go. That's why turning the rudder allows the submarine to change direction.

In the horizontal plane, when you turn the rudder, the submarine will rotate as it goes forward. It's like when you're driving a car, you're going forward and you're turning. At the same time, the submarine also creates something called a <b style="color:red">Drift Angle</b>, just like your car tilts when you turn.

On the vertical side, when you want the submarine to go up or down, you turn another thing called the elevator. It's like when you're driving a car, you want to speed up or slow down. Underwater, though, things get a little more complicated. Sometimes, even if you want a submarine to come up, it won't necessarily come up. This is because the relationship between the buoyancy of the water and the gravity of the submarine is complex. Similarly, when you turn the elevator, the submarine doesn't spin the same way you turn the rudder.

### 1.4 The Relation between Absolute Change and Relative Change of Vector

Now, imagine that you're looking at a vector $\vec{B}$ that changes over time. You can choose to observe it in two different coordinate systems: one is fixed (that is, the fixed coordinate system), and the other is moving (that is, the coordinate system with which the object is moving).

In a definite set, the change you see in vector $\vec{B}$ is called an absolute change, and it's denoted by $\Delta B$. In a dynamic system, the change you see in vector B is called a relative change and is represented by $\widetilde{\Delta B}$.

Generally, we called this vector $\vec{B}$ is Momentum Vector.

<center>
<img src="https://z1.ax1x.com/2023/10/04/pPOqYcR.png" width="42%"/>
</center>

Now, let's deduce the relation. Here, we need to emphasize that we assume that the moving system is moving around the circle of the fixed system with an angular velocity of $\Omega$:

$$
\Delta \vec{B}=\widetilde{\Delta B}+bc=\widetilde{\Delta B}+\vec\theta\times\vec{B} =\widetilde{\Delta B}+\vec\Omega \Delta t \times\vec B
$$

Therefore, divide both sides of the above equation by $\Delta t$, and then take the limit, we get:

$$
\lim_{\Delta t\to 0}\frac{\Delta\vec B}{\Delta t}=\lim_{\Delta t \to 0}\frac{\widetilde{\Delta B}}{\Delta t}+\lim_{\Delta t \to 0}(\vec\Omega \times \vec B)
$$

$$
\frac{\text{d}\vec{B}}{\text{d}t}=\frac{\text{d}\widetilde{B}}{\text{d}t}+\vec\Omega \times \vec B \tag{1}
$$

Similarly, we have a similar expression for the moment of momentum $\vec{K}$. However, it should be noted that our dynamic system may also be moving with the velocity $\vec{V}$ relative to the definite system, so we also need to add a term $\vec{V}\times\vec{B}$ at the end, which means:

$$
\frac{\text{d}\vec{K}}{\text{d}t}=\frac{\text{d}\widetilde{\vec{K}}}{\text{d}t}+\vec{\Omega}\times\vec{K}+\vec{V}\times\vec{B} \tag{2}
$$

### 1.5 General Equation of Submarine Motion

According to the content above, the following releation holds if the origin of dynamic system is the center of mass of the object:

$$
\frac{\text{d}\vec{B}}{\text{d}t}=\vec F
$$

$$
\frac{\text{d}\vec K}{\text{d}t}=\vec M
$$

where:
 - $\vec B$ -- Submarine Momentum
 - $\vec K$ -- The moment of momentum of the boat's motion with respect to the center of mass $G$
 
Plug the equation $(1)$ and $(2)$ into the expression above, we get:

$$
\vec{F} = \frac{\text{d}\vec{B}}{\text{d}t} = \frac{\text{d}\widetilde{B}}{\text{d}t}+\vec\Omega \times \vec B \tag{3}
$$

$$
\vec M = \frac{\text{d}\vec{K}}{\text{d}t}=\frac{\text{d}\widetilde{\vec{K}}}{\text{d}t}+\vec{\Omega}\times\vec{K}+\vec{V}\times\vec{B} \tag{4}
$$

Therefore, we need to compute the concrete expressions of the expressions $(3)$ and $(4)$, so we need to expand the cross product of the vectors.

Now assume that the components of $\vec B$ in the Cartesian coordinate system along the three orthogonal axes are: 

$$
B_x=mu \quad B_y=mv \quad B_z=mw \tag{5}
$$

For ease of calculation, we assume that $G-xyz$ is the central axis of inertia of the submarine (for non-approximations, use the parallel axis theorem to calculate the actual moment of inertia).

$$
K_x=I_x p \quad K_y=I_y q \quad K_z=I_z r \tag{6}
$$

So, the value of $\vec \Omega \times \vec{B}$ is:

$$
\begin{aligned}
\vec \Omega \times \vec{B}&=
\begin{vmatrix}
\vec{i} & \vec{j} & \vec{k} \\
p & q & r \\
B_x & B_y & B_z
\end{vmatrix}
 \\
&= (qB_z-rB_y)\vec{i}+(rB_x-pB_z)\vec{j}+(pB_y-qB_z)\vec{k}
\end{aligned} \tag{7}
$$

Therefore, by substituting the formula $(7)$ into the formula $(3)$ and the formula $(4)$, we get the following $6$ equations:

$$
\begin{cases}
m(\frac{\text{d}u}{\text{d}t}+qw-rv)=X \\
m(\frac{\text{d}v}{\text{d}t}+ru-pw)=Y \\
m(\frac{\text{d}w}{\text{d}t}+pv-qu)=Z \\
I_x\frac{\text{d}p}{\text{d}t}+(I_z-I_y)qr=K \\
I_y\frac{\text{d}q}{\text{d}t}+(I_x-I_z)rp=M \\
I_z\frac{\text{d}r}{\text{d}t}+(I_y-I_x)pq=N
\end{cases}
\tag{8}
$$

We call the above equation $(8)$. For convenience, we will add a dot to all the derivatives of time, for example: $\frac{\text{d}u}{\text{d}t} = \dot{u}$. Those above expressions are the general equation of submarine motion with $6$ degrees of freedom.

In those equations:
 - $(qw-rv), (ru-pw), (pv-qu)$ represensts the centripetal acceleration of the object due to the motion of the coordinate system;
 - $(I_z-I_y)qr, (I_x-I_z)rp, (I_y-I_x)pq$ represents gyration effect.
 
When studying the plane motion of submarine, generally do not consider the rolling surface motion, only consider the horizontal plane and vertical plane motion, their general equations of motion are:

$$
\text{Horizontal Plane} \ 
\begin{cases}
\begin{aligned}
m(\dot{u}-rv)&=X \\
m(\dot{v}+ru)&=Y \\
I_z\dot{r}&=N
\end{aligned}
\end{cases}\tag{9}
$$

$$
\text{Vertical Plane} \ 
\begin{cases}
\begin{aligned}
m(\dot{u}+qw)&=X \\
m(\dot{v}-qu)&=Z \\
I_y\dot{q}&=M
\end{aligned}
\end{cases}\tag{10}
$$

## 2 Inertial Hydrodynamic Coefficient

### 2.1 Fluid Inertia Force and Added Mass

The magnitude of the hydrodynamic force experienced by an object in an unsteady motion in an ideal fluid is proportional to the acceleration of the object's motion and the direction is opposite to the acceleration direction, while the proportionality constant is called the additional mass and is represented by $\lambda_{ij}$. It is when moving with unit (angular) acceleration in the direction $i$, the additional mass, the additional static moment of mass, and the additional moment of inertia in the direction $j$, are the hydrodynamic inertia forces experienced by an object moving with unit (angular) acceleration in an ideal fluid.

The convention are as follows:

The movement along the x, y, and z directions is denoted by $1$, $2$, and $3$;

Rotation along the x, y, and z directions is represented by $4$, $5$, and $6$;

Therefore, an object of any shape moves with a total of $36$ additional masses:

$$
\begin{bmatrix}
\lambda_{11} & \lambda_{12} & \lambda_{13} & \lambda_{14} & \lambda_{15} & \lambda_{16} \\
\lambda_{21} & \lambda_{22} & \lambda_{23} & \lambda_{24} & \lambda_{25} & \lambda_{36} \\
\lambda_{31} & \lambda_{32} & \lambda_{33} & \lambda_{34} & \lambda_{35} & \lambda_{36} \\
\lambda_{41} & \lambda_{42} & \lambda_{43} & \lambda_{44} & \lambda_{45} & \lambda_{46} \\
\lambda_{51} & \lambda_{52} & \lambda_{53} & \lambda_{54} & \lambda_{55} & \lambda_{56} \\
\lambda_{61} & \lambda_{62} & \lambda_{63} & \lambda_{64} & \lambda_{65} & \lambda_{66} 
\end{bmatrix}
$$

From the conclusion of fluid dynamics:

$$
\lambda_{ij}=\lambda_{ji} \quad (i,j=1,2,\dots,6)
$$

The inertia force is:

$$
R_j=-\lambda_{ij}\dot{v_i}\tag{11}
$$

The added mass is frequency dependent as the ship moves in waves. For submarine maneuverability, however, the change in motion is slow and the effect of frequency can be ignored, i.e. the additional mass of zero frequency.

According to equation $(11)$, we can get the equation in scaler form:

$$
R_x=-\lambda_{11}\dot{v_1} \\
R_y=-\lambda_{12}\dot{v_1} \\
R_z=-\lambda_{13}\dot{v_1}
$$

The torque of force $R$ on the three coordinate axes is:

$$
m_x=-\lambda_{14}\dot{v_1} \\
m_y=-\lambda_{15}\dot{v_1} \\
m_z=-\lambda_{16}\dot{v_1} 
$$

For a general hull, there is a symmetry plane $xOz$. Therefore, when the object moves parallel to the symmetry plane, the surrounding flow field is also symmetric to the $xOz$ plane. There is a fluid pressure $p$ on the right microarea element $ds_1$, then there must be a symmetric pressure $p$ on the left symmetrical microarea element $ds_2$, and the resultant fluid force R of the whole object must be located in the symmetric plane $xOz$. Therefore, the projection of force $R$ on the oy axis is zero, and the torque of force $R$ on the ox and oz axes is also zero. So:

$$
\begin{cases}
\lambda_{12}=\lambda_{14}=\lambda_{16}=0 \\
\lambda_{32}=\lambda_{34}=\lambda_{36}=0 \\
\lambda_{52}=\lambda_{54}=\lambda_{56}=0
\end{cases}\tag{12}
$$

As a result:

$$
\begin{bmatrix}
\lambda_{11} & 0 & \lambda_{13} & 0 & \lambda_{15} & 0 \\
0 & \lambda_{22} & 0 & \lambda_{24} & 0 & \lambda_{36} \\
\lambda_{31} & 0 & \lambda_{33} & 0 & \lambda_{35} & 0 \\
0 & \lambda_{42} & 0 & \lambda_{44} & 0 & \lambda_{46} \\
\lambda_{51} & 0 & \lambda_{53} & 0 & \lambda_{55} & 0 \\
0 & \lambda_{62} & 0 & \lambda_{64} & 0 & \lambda_{66} 
\end{bmatrix}
$$

By looking at the matrix above, if the object has a plane of symmetry, then the value of the term where the sum of the corner signs of $\lambda$ is odd is $0$.

### 2.2 Inertial Hydrodynamics acting on the Body of the Boat

According to fluid mechanics, when an object moves in an infinite ideal fluid, the kinetic energy of the fluid disturbance motion is:

$$
T=\frac{1}{2}\sum_{i=1}^{6}\sum_{j=1}^{6}\lambda_{ij}v_iv_j \quad (i,j=1,2,\dots,6)
$$

The momentum $B$ of the fluid disturbance motion is related to the kinetic energy $T$ as follows:

$$
B_i=\frac{\partial T}{\partial V_i}\quad (i=1,2,\dots,6)
$$

Plug in, and apply the values of $(12)$:

$$
\begin{aligned}
B_1&=B_x=\frac{\partial T}{\partial u}=\lambda_{11}u \\
B_2&=B_y=\frac{\partial T}{\partial v}=\lambda_{22}v+\lambda_{24}p+\lambda_{26}r \\
B_3&=B_z=\frac{\partial T}{\partial w}=\lambda_{33}w+\lambda_{35}q \\
B_4&=K_x=\frac{\partial T}{\partial p}=\lambda_{44}p+\lambda_{46}r+\lambda_{24}v \\
B_5&=K_y=\frac{\partial T}{\partial q}=\lambda_{55}q+\lambda_{35}w \\
B_6&=K_z=\frac{\partial T}{\partial r}=\lambda_{66}r+\lambda_{46}p+\lambda_{26}v 
\end{aligned}
$$

According to the relationship:

$$
\begin{cases}
F_I=-\frac{\text{d}\vec B}{\text{d} t} \\
M_I=-\frac{\text{d}\vec K}{\text{d} t}
\end{cases}
$$

So:

$$
\begin{aligned}
X_I&=-\lambda_{11}\dot{u}-\lambda_{35}q^2+\lambda_{26}r^2+\lambda_{22}vr-\lambda_{33}wq+\lambda_{24}pr \\
Y_I&=-\lambda_{22}\dot{v}-\lambda_{24}\dot{p}-\lambda_{26}\dot{r}-\lambda_{11}ur+\lambda_{33}wp+\lambda_{35}pq \\
Z_I&=-\lambda_{33}\dot{w}-\lambda_{35}\dot{q}-\lambda_{24}p^2+\lambda_{11}uq-\lambda_{22}vp-\lambda_{26}pr \\
K_I&=-\lambda_{24}\dot{v}-\lambda_{44}\dot{p}-\lambda_{46}\dot{r}+(\lambda_{22}-\lambda_{33})vw-(\lambda_{25}+\lambda_{26})vq+\lambda_{24}wp+(\lambda_{26}+\lambda_{35})wr-\lambda_{46}pq+(\lambda_{55}-\lambda_{66})qr \\
M_I&=-\lambda_{35}\dot{w}-\lambda_{55}\dot{q}+(\lambda_{33}-\lambda_{11})uw+\lambda_{35}uq+\lambda_{46}p^2-\lambda_{46}r^2+\lambda_{26}vp-\lambda_{24}vr+(\lambda_{66}-\lambda_{44})pr \\
N_I&=-\lambda_{26}\dot{v}-\lambda_{46}\dot{p}-\lambda_{66}\dot{r}+(\lambda_{11}-\lambda_{22})uv-\lambda_{24}up-\lambda_{26}ur+\lambda_{24}vq-\lambda_{35}wp+(\lambda_{44}-\lambda_{55})pq+\lambda_{46}qr
\end{aligned} 
\tag{13}
$$

For the planner motion, we apply the following relation:

$$
w=q=\dot w=\dot q=p=\dot p=0
$$

then, the expression $(13)$ can be simplfied to:

$$
\begin{cases}
X_I=-\lambda_{11}\dot{u}+\lambda_{22}vr+\lambda_{26}r^2 \\
Y_I=-\lambda_{22}\dot{v}-\lambda_{11}ur-\lambda_{26}\dot{r} \\
N_I=-\lambda_{66}\dot{r}+(\lambda_{11}-\lambda_{22})uv-\lambda_{26}(\dot v+ur)
\end{cases}
$$

### 2.3 The corresponding relationship between the additional mass and the acceleration coefficient

According to the meaning of add-mass entry, subscripts $i$ and $j$, it is compared with the hydrodynamic coefficients of each (angular) acceleration:

For example:

The inertial force due to $\dot u$ is:

$$
X(\dot u)=X_{\dot u}\dot u
$$

The longitudinal component of the fluid inertia force expressed in $\lambda$ is:

$$
X_I=-\lambda_{11}\dot{u}
$$

So by observing, the term $\lambda_{11}$ and $X_{\dot u}$ have same value, but only have difference in sign, so similarly:

$$
\begin{cases}
\lambda_{11}=-X_{\dot u} \\
\lambda_{22}=-Y_{\dot v} \\
\lambda_{33}=-N_{\dot r} \\
\lambda_{26}=-Y_{\dot r}=-N_{\dot v} \\
\lambda_{33}=-Z_{\dot w} \\
\lambda_{55}=-M_{\dot q} \\
\lambda_{35}=-Z_{\dot q}=-M_{\dot w}
\end{cases}
$$

So you take the acceleration coefficient that doesn't have dimensionality and you divide the force by $\frac{1}{2}\rho V^2 L^2$, and you divide the moment by $\frac{1}{2}\rho V^2 L^3$, and you don't dimensionalize them, and then you use a prime to represent the dimensionality. So, we get the following:

$$
\begin{cases}
X_{\dot u}'=\frac{X_{\dot u}}{\frac{1}{2}\rho V^2 L^3} \\
Y_{\dot v}'=\frac{Y_{\dot v}}{\frac{1}{2}\rho V^2 L^3} \\
Z_{\dot w}'=\frac{Z_{\dot w}}{\frac{1}{2}\rho V^2 L^3} \\
Y_{\dot r}'=N_{\dot v}'=\frac{Y_{\dot r}}{\frac{1}{2}\rho L^4}
=\frac{N_{\dot v}}{\frac{1}{2}\rho L^4} \\
Z_{\dot q}'=M_{\dot w}'=\frac{Z_{\dot q}}{\frac{1}{2}\rho L^4}
=\frac{M_{\dot w}}{\frac{1}{2}\rho L^4} \\
N_{\dot r}'=\frac{N_{\dot r}}{\frac{1}{2}\rho L^5} \\
M_{\dot q}'=\frac{M_{\dot q}}{\frac{1}{2}\rho L^5}
\end{cases}
$$

and

$$
\dot{u}'=\frac{\dot{u}L}{V^2} \quad \dot{v}'=\frac{\dot{v}L}{V^2} \quad \dot{w}'=\frac{\dot{w}L}{V^2} \\
\dot{q}'=\frac{\dot{q}L}{V^2} \quad \dot{r}'=\frac{\dot{r}L}{V^2}
$$

where $L$ is the length; $\rho$ is the density of water; $V$ is speed.

## 3 The Viscous Hydrodynamic Force

For a certain submarine moving in infinitely deep, wide and still water, we can consider $\dot{u}=\dot{v}=\dot{r}=\dot{w}=\dot{q}=0$ to be true, then:

$$
\begin{cases}
X=f_X(u,v,r,\delta \\
Y=f_Y(u,v,r,\delta \\
Z=f_Z(u,v,r,\delta
\end{cases}
$$

and

$$
X=f_X(u,w,q,\delta_b,\delta_s) \\
Z=f_Z(u,w,q,\delta_b,\delta_s) \\
W=f_W(u,w,q,\delta_b,\delta_s)
$$



### 3.1 Viscous Hydrodynamic Forces Acting on a Horizontal Moving Submarine

It is well known that submarines have a symmetrical shape, so when $v,r,\delta$ changes sign, $X$ is completely unchanged, so you can safely and boldly assume that it is an even function of $v,r,\delta$. Since the expression of $X$ does not contain odd terms such as $v,v^3,r,r^3,v^2r,vr^2$, the coefficients of these terms such as $\frac{\partial X}{\partial r}, \frac{\partial^3 X}{\partial r^3},\dots, \frac{\partial^3 X}{\partial r \partial r^2}$ are $0$.

<center>

Similarly, when the sign of $v,r,\delta$ changes, the absolute value of $Y$ and $N$ does NOT change. That is, $Y$ and $N$ are odd-function with respect to $v$, $r$ and $\delta$. Therefore, the even-order derivatives of Y, N with respect to $v, r$, and $\delta$ and the even-order coupling derivatives are zero.

For the study of submarine's course stability and maneuvering motion under small rudder Angle, the deviation of motion parameters from the initial reference motion state of $\Delta u$, $v$, $r$, $\delta$is small, and the second and above terms in the hydrodynamic Taylor expansion can be ignored, and only the linear terms are retained. Thus, the linear expression of the hydrodynamic force can be obtained as follows:

$$
\begin{cases}
X=X_0+X_u\Delta u \\
Y=Y_0+Y_v\Delta u+Y_vv+Y_rr+Y_\delta\delta \\
N=N_0+N_w\Delta u+N_vv+N_rr+N_\delta\delta
\end{cases}
$$

Similarly, the change of the speed $\Delta u$ in the $x$ direction will not cause the lateral force $Y_u\Delta u$ and the torque $N_u\Delta u$, that is, $Y_u=N_u=0$, then:

$$
\begin{cases}
X=X_0+X_u\Delta u \\
Y=Y_vv+Y_rr+Y_\delta\delta \\
N=N_vv+N_rr+N_\delta\delta
\end{cases}
$$

Perform dimensionless operations:

$$
X'=\frac{X}{\frac{1}{2}\rho V^2L^2} \\
Y'=\frac{Y}{\frac{1}{2}\rho V^2L^2} \\
N'=\frac{N}{\frac{1}{2}\rho V^2L^3} \tag{14}
$$

The formulaes $(14)$ can be rewritten dimensionless as:

$$
\begin{cases}
X'=X_0'+X_u'\Delta u' \\
Y'=Y_v'v'+Y_r'r'+Y_\delta'\delta' \\
N'= N_v'v'+N_r'r'+N_\delta'\delta'
\end{cases}
$$

However, what we need to pay attention to is: In order to study the maneuvering motion of submarine under large rudder Angle and the maneuverability of ship with unstable course, <b style="color:red">the nonlinear term of motion parameters must be considered</b>.

### 3.2 Viscous hydrodynamic forces acting on a vertical motion submarine

#### 3.2.1 Linear and Non-Linear Expressions

When the direction of $w$ is changed, the hydrodynamic force caused by positive and negative Angle of attack a is different, and the change law of $Z(+w)$ and $Z(-w)$ is also different:

$$
Z(w)=Z_ww+Z_{|w|}|w|
$$

where:
 - $Z_w$ represents the mean velocity coeffecient in the area of $\pm w$
 - $Z_{|w|}$ represents the velocity coefficient correction for boat asymmetry
 
Similarly, the linear hydrodynamic expression of the pitching moment $M(w)$ is:

$$
M(w)=M_ww+M_{|w|}|w|
$$

where:
 - $M_w$ represents the mean velocity coeffecient
 - $M_{|w|}$ represents the correction value of the velocity coefficient, which is zero when $M(w)$ is a symmetric function

The upper and lower dissymmetry of the hull will have a similar effect on other hydrodynamic forces. However, this treatment method greatly complicates the hydrodynamic expression, and the effect of this asymmetry is still small. **Therefore, the remaining hydrodynamic coefficients no longer take into account the asymmetry correction**.

#### 3.2.2 The meaning and symbol of hydrodynamic derivative

##### 3.2.2.1 About zero lift $Z_0$ and zero lift moment $M$

The following relation holds if the submarine under this condition:

$$
u=u_0, w=q=\dot u=\dot w=\dot q=\delta_b=\delta_s=0
$$

As a result:

$$
\begin{cases}
Z_0=Z_0'·\frac{1}{2}\rho V^2L^2 \\
M_0=M_0'·\frac{1}{2}\rho V^2L^3
\end{cases}
$$

##### 3.2.2.2 Position derivative related content
 
The derivatives $Z_w$ and $M_w$ are the slopes of the vertical force $Z(w)$ and trim force $M(w)$ curves at the origin $(w=0)$ when the submarine is moving with longitudinal velocity $u$ and vertical velocity $w$.

##### 3.2.2.3 Elevator rudder Angle derivative related content

$$
Z_{\delta_b}<0, \ M_{\delta_b}>0 \\
Z_{\delta_s}<0, \ M_{\delta_s}<0 \\
Z_{q}<0 \ \text{or} \ Z_{q}>0 \ M_q<0 \\
$$

These hydrodynamic coefficients (derivatives) are a constant for a certain submarine, which can be determined by experimental and theoretical methods, and are mainly obtained by constrained ship model tests in engineering.


## 4 Equation of Motion for Horizontal Maneuvering

By substituting the hydrodynamic representation of submarine into the general equation of submarine maneuvering motion, the submarine maneuvering side can be obtained.

### 4.1 Linear equation of Horizontal Maneuvering Motion

The linear equation is only suitable for weak maneuvers with small motion parameters $\Delta u, v,r,\delta$:

$$
rv=\Delta r·\Delta v \approx 0
ur=(u_0+\Delta u)\Delta r=u_0\Delta r+\Delta u·\Delta r \approx u_0 r
$$

So:

$$
\begin{cases}
m\dot u=X \\
m(\dot v+u_0 r)=Y \\
I_z\dot r=N
\end{cases} \tag{15}
$$

Note:
 - The $2^{\text{nd}}$ and $3^{\text{rd}}$ expression in equations $(15)$ not contain $u$, but only $u_0$, so it can be seperated from $1^{\text{st}}$ expression.
 - The $1^{\text{st}}$ expression in $(15)$ is used to determine the velocity $v$. In the weak maneuver, $u$ changes very little, generally $u=u_0$.
 - When the hydrodynamic function of the submarine is expanded by Taylor series, the second order and the above terms in the expansion formula are omitted, and only the linear terms are retained.
 
Then:

$$
\begin{cases}
(m-Y_{\dot v})\dot{v}-Y_vv+Y_{\dot r}\dot{r}+(mV-Y_r)r=Y_\delta\delta \\
(I_z-N_{\dot r})\dot{r}-N_rr-N_{\dot v}\dot v-N_vv=N_\delta\delta
\end{cases} 
$$

This is called the **linear equation of submarine horizontal maneuvering motion**.

### 4.2 Nonlinear Equation of Horizontal Maneuvering Motion

The actual motion of submarine shows that the values of motion parameters $v$ and $r$ are large, and the nonlinear effects of velocity and hydrodynamic dynamics must be considered.

$$
\begin{cases}
m(\dot{u}-vr)=X_{\dot u}\dot{u}+X_{uu}u^2+X_{vv}v^2+X_{rr}r^2+X_{vr}vr+X_{\delta\delta}\delta^2+X_T \\
m(\dot{v}+ur)=Y_{\dot{v}}\dot{v}+Y_{\dot{r}}\dot{r}+Y_vv+Y_rr+Y_{v|v|}v|v|+Y_{v|r|}v|r|+Y_{\delta}\delta+Y_{|r|\delta}|r|\delta \\
I_z\dot{r}=N_{\dot{v}}\dot{v}+N_{\dot{r}}\dot{r}+N_vv+N_rr+N_{v|v|}v|v|+N_{v|r|}v|r|+N_{r|r|}r|r|+N_{\delta}\delta+N_{|r|\delta}|r|\delta
\end{cases}
$$

## 5 Equation of Motion for Vertical Control

### 5.1 Linear equation of maneuvering motion in vertical plane

When the submarine makes a weak maneuver in the vertical plane, imitating the linear equation of the horizontal plane, there are also:

$$
\begin{cases}
m\dot{u}=X \\
m(\dot{w}-u_0q)=Z \\
I_y\dot{q}=M
\end{cases}
$$

Similarly:

$$
\begin{cases}
(m-Z_{\dot w})\dot{w}-Z_{w}-Z_{\dot{q}}\dot{q}-(mV+Z_q)q
=Z_0+Z_{\delta_b}\delta_b+Z_{\delta_s}\delta_s+P \\
(I_y-M_{\dot q})\dot q-M_qq-M_{\dot w}\dot w-M_ww=M_0+M_{\delta_b}\delta_b+M_{\delta_s}\delta_s+X_Tz_T+M_p+M_\theta\theta
\end{cases}
$$

### 5.2 Vertical Plane Manipulation Nonlinear Equation

$$
\begin{cases}
m(\dot{u}+wq)=X_{\dot u}\dot u+X_{uu}u^2+X_{ww}w^2+X_{qq}q^2+X_{wq}wq+X_{\delta_s\delta_s}\delta_s\delta_s+X_{\delta_b\delta_b}\delta_b\delta_b+X_T+P\theta \\
m(\dot{w}-uq)=Z_0+Z_{\dot w}\dot w+Z_{\dot q}\dot q+Z_ww+Z_{|w|}|w|+Z_qq+Z_{w|w|}w|w|+Z_{ww}w^2+Z_{|w|q}w|q|+Z_{q|q|}q|q|+Z_{\delta_s}\delta_s+Z_{\delta_b}\delta_b+Z_{|q|\delta_s}|q|\delta_s+P \\
I_y\dot{q}=M_0+M_{\dot w}\dot w+M_{\dot q}\dot q+M_ww+M_{|w|}|w|+M_qq+M_{w|w|}w|w|+M_{ww}w^2+M_{w|q|}w|q|+M_{q|q|}q|q|+M_{\delta_s}\delta_s+M_{\delta_b}\delta_b+M_{|q|\delta_s}|q|\delta_s+X_TZ_T+M_p+M_\theta\theta \\
q=\frac{\text{d}\theta}{\text{d}t}=\dot{q}
\end{cases}
$$

The above motion equation contains four unknown motion parameters, $u(t), w(t), q(t)$, and $\theta(t)$, usually only the axial force equation ($X$ equation), vertical force equation ($Z$ equation) and trimming moment equation ($M$ equation), to supplement the motion relation $q=\dot\theta$.

When the control rule function is given, the motion response function of the submarine can be solved.

## 6 Summary

Ship maneuverability equation is an important tool in navigation, which is used to describe the motion characteristics and maneuverability of ships under various environmental conditions. These equations help us understand how the ship responds to inputs from rudders, propellers, and other control devices, as well as how the ship interacts with the surrounding current.


The significance of ship maneuverability equation lies in the fact that it can accurately simulate the dynamic behavior of ship in the course of sailing. Through these equations, we can predict and optimize the trajectory, speed and attitude of the ship under various sailing conditions to achieve safe, efficient and energy efficient sailing. In addition, these equations also provide an important basis for ship design, construction and retrofit, and help engineers understand the dynamic characteristics of ships and carry out targeted optimization.


In short, ship maneuverability equation provides us with a key tool to understand and predict ship behavior, which is of great significance for Marine safety, efficiency and energy saving.
